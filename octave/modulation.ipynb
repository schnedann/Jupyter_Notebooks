# Modulation


```octave
clear all;
close all;

fig=1;

#available_graphics_toolkits()
#loaded_graphics_toolkits()
graphics_toolkit ("gnuplot");
```

# discrete Signal

```octave
fsampl = 25e6; #[Hz]
t_end = 0.01; #[s]
t= linspace(0,t_end,t_end.*fsampl);
```

## carrier

```octave
fT = 10e6; #[Hz]
phase = (3.*pi)./4; #[rad] - 135
x1 = 2.*pi.*t.*fT+phase;
```

## signal

```octave
fS = 3e6; #[Hz]
x2 = 2.*pi.*t.*fS;
```

## modulated signal

```octave
am_signal = sin(x1).*sin(x2);
```

```octave
figure(fig++);
plot(t,am_signal);
grid on;
```

```octave
Fs=20.*log10(abs(fft(am_signal)));
figure(fig++);
freqencies=linspace(0,fsampl./2,length(Fs)./2);
plot(freqencies,Fs(:,1:1:(length(Fs)./2)));
grid on;
```
